### Question 1


**What is the most common trip duration under an hour for New York vs. Chicago?.**

```r
ny$city = "New York" # add column name with city for histogram
head(ny)
chi$city = "Chicago"
head(chi)
ny.chi = rbind(ny, chi) # combine datasets
head(ny.chi) # checked to see both datasets present
tail(ny.chi)

library(ggplot2)
by(ny.chi$Trip.Duration/3600, ny.chi$city, summary) # checked summary data and realized the distrubtion was too extreme

# changed seconds to hours by dividing by 3,600 and omitted rows without end station
ggplot(aes(x = Trip.Duration/3600), data = subset(ny.chi, !is.na(Trip.Duration))) +
  geom_histogram(bins = 30, color = 'black', fill = '#F79420' ) +
  xlim(0,1) +
  xlab('Trip Duration Over an Hour') +
  ylab('Number of Bike Rentals') +
  ggtitle('Bike Use Frequency Overr an Hour - New York vs. Chicago') +
  facet_wrap(~city)
```

**Looking at these histograms there is a higher frequency of usage in New York than Chicago over the period of an hour of usage. Both cities have similar histograms with frequency being the highest around .13 hours and tapering off as one hour approaches. Shorter periods of usage are more frequent for both cities when looking at an hour time frame. My original question took into account all hours of usage, but there were so many outliers (such as usage of over 25 days) I altered my original question to include more frequent usage. I removed any data that did not show an end station, becuase I felt it was not good quality data.**

### Question 2

**What is the mean time usage (under an hour of usage) by age for New York and how is it distrubuted?**

```r
ny$age <- 2017 - ny$Birth.Year # calculating age subtracting year data created by birth year
ny.1 <- subset(ny, !is.na(age)) # removing NA 
head(ny)

# plotted mean over scatter plot, used jitter to smooth data
# set limits on age based on outliers for older riders that was suspicious and ack of data for younger riders
# set limit to one hour based on distribution
# removed rows of data with no birth year
ggplot(aes(x = age, y = Trip.Duration/3600), data = ny) +
  geom_jitter(alpha = 1/20, color = "orange") +
  ylim(0,1) +
  xlim(18, 80) +
  geom_line(stat = 'summary', fun.y=mean) +
  xlab('Bike Rider Age') +
  ylab('Time Duration') +
  ggtitle('New York Bike Rider Usage Under an Hour by Age')
```

**The mean was evenly distrubted among age groups for users renting bikes under one hour. The mean is a little under 0.25 hours. A majority of users are 18 - 60 years old with more users being under 40.**

### Question 3

**What is the average and most common age for riders in New York versus Chicago from 18 to 80 years old?**

```r
ny.chi$age <- 2017 - ny.chi$Birth.Year # calculating age subtracting year data created by birth year
ny.chi.1 <- subset(ny.chi, !is.na(age)) # remove rows with no age present
head(ny.chi.1)
ny.chi.1 <- subset(ny.chi.1, age <= 80 | age >= 18) # limiting age from 18 - 80
head(ny.chi.1)
by(ny.chi.1$age, ny.chi.1$city, summary) 

ggplot(aes(x= city, y = age), data = ny.chi.1) +
  geom_boxplot() +
  ylim(18,80) +
  xlab('City') +
  ylab('Age') +
  ggtitle('Bike Rider Age by City') 
```

**The most common age for riders in Chicago was 33 and New York was 36 years old. The means were also similar with Chicago having an average age of 36.04 and New York was 38.82. I removed some outliers and limited the age from 18 - 80**
